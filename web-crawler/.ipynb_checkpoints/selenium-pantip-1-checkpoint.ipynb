```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://pantip.com/forum/supachalasai")
```

```python
for i in range(10):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
```

```python
data = driver.find_elements_by_class_name('post-item-title')
```

```python
for _d in data:
    print(_d.text)
```
